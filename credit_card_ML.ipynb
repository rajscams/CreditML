```python
card_df = pd.read_csv(r"C:\Users\Surface\Downloads\credit_card_data\Credit_card.csv")
target_df = pd.read_csv(r"C:\Users\Surface\Downloads\credit_card_data\Credit_card_label.csv")
```

```python
card_df = card_df.drop("Birthday_count", axis=1)
card_df = card_df.drop("Work_Phone", axis=1)
card_df = card_df.drop("Mobile_phone", axis=1)
card_df = card_df.drop("EMAIL_ID", axis=1)
card_df = card_df.drop("Phone", axis=1)
card_df = card_df.drop("Type_Occupation", axis=1)
```

```python
card_df = card_df.dropna()
```

```python
card_df.isna().sum()
```

```python
def hot_encoder(df, column):
    df_encoded = pd.get_dummies(df[column], dtype= int)
    df = df.join(df_encoded)
    df = df.drop(column, axis=1)
    return df
```

```python
card_df = hot_encoder(card_df, "GENDER")
```

```python
card_df["Car_Owner"] = card_df.Car_Owner.replace(('Y', 'N'), (1,0))
```

```python
card_df["Property_Owner"] = card_df.Propert_Owner.replace(('Y', 'N'), (1,0))
```

```python
card_df = card_df.drop('Propert_Owner', axis=1)
card_df.head()
```

```python
education_list = card_df.EDUCATION.unique()
education_list
```

Higher Education and Academic degree is consider to be the same. Data is ordinal so enconding will have an hierarchy.

```python
card_df = card_df.replace(education_list, [4,2,1,3,4])
```

```python
card_df
```

```python
card_df = hot_encoder(card_df, 'Marital_status')
```

```python
card_df = card_df.drop('Type_Income', axis=1)
```

```python
card_df
```

```python
merged_df = pd.merge(card_df,target_df, on='Ind_ID')
```

```python
merged_df
```

```python
merged_df = merged_df.drop('Ind_ID', axis=1)
merged_df = merged_df.drop('Housing_type', axis=1)
```

```python
merged_df.corr()
```

```python
import sklearn 
from sklearn.model_selection import train_test_split

Y = merged_df['label']
X = merged_df[['Car_Owner', 'CHILDREN', 'Annual_income', 'EDUCATION', 'Employed_days', 'Family_Members','F', 'M', 'Property_Owner', 'Civil marriage', 'Married', 'Separated','Single / not married','Widow']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=42, shuffle=True, test_size=0.3)

# Confirm how the data was split
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier

Pipeline_lr = Pipeline([('lr_classifier', LinearRegression())])
Pipeline_rf = Pipeline([('lr_classifier', RandomForestClassifier(n_estimators=10, random_state=10))])

Pipelines = [Pipeline_lr,Pipeline_rf]
pipe_dict = {0: 'Linear Regression', 1: 'Random Forest Regressor'}

for pipe in Pipelines:
    pipe.fit(X_train, y_train)

for i,model in enumerate(Pipelines):
    pred_test = model.predict(X_test)
    if i == 1: rf_model = model
    print("{} Test Accuracy:{}".format(pipe_dict[i],model.score(X_test,y_test)))
    print(pd.DataFrame({'Actual': y_test, 'Predicted': pred_test}))
    
```
